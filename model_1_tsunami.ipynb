```python
import numpy as np
import pandas as pd
import tensorflow as tf
from google.colab import files

uploaded = files.upload()

df = pd.read_csv('wave_data.csv')
df
```

```python
entrada = tf.keras.layers.Dense(units = 3, input_shape=[1])
oculta1 = tf.keras.layers.Dense(units = 3)
salida = tf.keras.layers.Dense(units = 1)

model = tf.keras.Sequential([entrada, oculta1, salida])
```

```python
#Compilamos el modelo

model.compile(
    optimizer = tf.keras.optimizers.Adam(0.1),
    loss = 'mean_squared_error'
)
```

```python
from re import X
#entrenamiento del modelo
x = df[["Pendiente (grados) "]]
y = df[["Posicion final de la costa"]]

x
```

```python
y
```

```python
print('Entrenando...')
historial = model.fit(x,y,epochs=200, verbose=False)
print('Modelo entrenado!')
```

```python
import matplotlib.pyplot as plt

plt.plot(historial.history['loss'])
plt.xlabel('cada epoca')
plt.ylabel('perdida')
plt.show()
```

```python
print('Hagamos una prediccion!')
input_data = np.array([[68]])
resultado = model.predict(input_data)
print('El resultado es ' + str(resultado) + ' metros de posicion final')
```
